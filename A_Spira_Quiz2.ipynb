```python
#We can see there are a lot of outliers
#We can replace the outliers with the median of the data
df_outlier = autodf.copy()

#let's check if any data is outside the IQR
IQR = df_outlier['engine-size'].quantile(.75) - df_outlier['engine-size'].quantile(.25)

#create a column in the data to report all instances of being outside the IQR
df_outlier['engine_size_outlier'] = False

#loop through the data checking what is outside the range
for index, row in autodf.iterrows():
  if row['engine-size'] > (df_outlier['engine-size'].quantile(.25)+1.5*IQR):
    df_outlier.at[index,'engine_size_outlier'] = True

outliers = df_outlier.loc[df_outlier['engine_size_outlier'] == True]
print(outliers)

#create a cache of the not-outlier data
non_outliers = df_outlier.loc[df_outlier['engine_size_outlier'] == False]

#get the median
median = non_outliers['engine-size'].median()
print("The median is ", median)

#loop through and set every outlier to the median
for index, row in df_outlier.iterrows():
  if row['engine_size_outlier'] == True:
    df_outlier.at[index,'engine-size'] = median

#drop the last column
df_outlier.drop('engine_size_outlier', axis = 1,inplace=True)

#show the sample
df_outlier.sample(7)

sns.boxplot(df_outlier['engine-size'])

```

```python
sns.histplot(df_outlier['engine-size'])
```

This method definately improved the data and removed a nice amount of the outliers. However, it is not perfect because some outliers are still there. In the histogram we can see that the furthest outliers were removed. Maybe the closer outliers were not picked up to be outliers and it is just more obvious on the box plot.

```python
from sklearn.preprocessing import Binarizer
binarizer = Binarizer(threshold=0)
binarized_features = binarizer.fit_transform(autodf[['symboling']])
print(binarized_features)
```

```python
from sklearn.preprocessing import LabelEncoder
X = autodf.drop('make',1)
y = autodf['make']

X = X.apply(LabelEncoder().fit_transform)
print(X.head())

#chi-square analysis
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, y)


feature_score = pd.DataFrame()

for i in range(X.shape[1]):
    new = pd.DataFrame({'Features' : X.columns[i],
                        'Score' : fit.scores_[i]}, index=[i])
    
    feature_score = pd.concat([feature_score, new])


#we will select those 
feature_score
```

```python
X_new = fit.transform(X)
X_new = pd.DataFrame(X_new)

X_new.head()
```
